```python
import pandas as pd

# Read the CSV file with the specified columns
df = pd.read_csv("merged_data.csv", usecols=['Col1', 'batter', 'total_runs_delivery', 'balls_faced', 'winning_team'])

# Identify the boundaries of each inning
inning_boundaries_indices = df[df['Col1'] == 0].index

# Append the end of the DataFrame as the last boundary
inning_boundaries_indices = list(inning_boundaries_indices) + [len(df)]

print("Inning Boundaries Indices:")
print(inning_boundaries_indices)

# Function to process each inning
def process_inning(df_inning):
    # Calculate the total runs
    total_runs = df_inning['total_runs_delivery'].sum()
    # Calculate the number of overs
    num_overs = (df_inning['Col1'].max() // 6) + 1
    # Calculate the average runs per over
    avg_runs_per_over = total_runs / num_overs
    return total_runs, num_overs, avg_runs_per_over

# List to store the results for each inning
inning_stats = []

# Iterate through each inning using the boundaries
for i in range(len(inning_boundaries_indices) - 1):
    start_idx = inning_boundaries_indices[i]
    end_idx = inning_boundaries_indices[i + 1]
    df_inning = df.iloc[start_idx:end_idx]
    total_runs, num_overs, avg_runs_per_over = process_inning(df_inning)
    inning_stats.append((total_runs, num_overs, avg_runs_per_over))

# Display the total runs, number of overs, and average runs per over for each inning
for i, (total_runs, num_overs, avg_runs_per_over) in enumerate(inning_stats, start=1):
    print(f"Inning {i}: Total Runs = {total_runs}, Number of Overs = {num_overs}, Average Runs per Over = {avg_runs_per_over:.2f}")
```

```python
import pandas as pd

# Read the CSV file with the specified columns
df = pd.read_csv("merged_data.csv", usecols=['Col1', 'batter', 'total_runs_delivery', 'balls_faced', 'winning_team'])

# Identify the boundaries of each inning
inning_boundaries_indices = df[df['Col1'] == 0].index

# Append the end of the DataFrame as the last boundary
inning_boundaries_indices = list(inning_boundaries_indices) + [len(df)]

print("Inning Boundaries Indices:")
print(inning_boundaries_indices)

# Function to process each inning
def process_inning(df_inning):
    # Calculate the total runs
    total_runs = df_inning['total_runs_delivery'].sum()
    # Calculate the number of overs
    num_overs = (df_inning['Col1'].max() // 6) + 1
    # Calculate the average runs per over
    avg_runs_per_over = total_runs / num_overs
    # Determine if the batting team is the winning team
    batting_team = df_inning['batter'].iloc[0]  # Assume the first batter represents the batting team
    winning_team = df_inning['winning_team'].iloc[0]
    is_batting_team_winning = batting_team == winning_team
    return total_runs, num_overs, avg_runs_per_over, is_batting_team_winning

# List to store the results for each inning
inning_stats = []

# Iterate through each inning using the boundaries
for i in range(len(inning_boundaries_indices) - 1):
    start_idx = inning_boundaries_indices[i]
    end_idx = inning_boundaries_indices[i + 1]
    df_inning = df.iloc[start_idx:end_idx]
    total_runs, num_overs, avg_runs_per_over, is_batting_team_winning = process_inning(df_inning)
    inning_stats.append({
        'Inning': i + 1,
        'Total Runs': total_runs,
        'Number of Overs': num_overs,
        'Average Runs per Over': avg_runs_per_over,
        'Batting Team Wins': is_batting_team_winning
    })

# Create a DataFrame from the list of dictionaries
inning_stats_df = pd.DataFrame(inning_stats)

# Display the DataFrame
print(inning_stats_df)
```
