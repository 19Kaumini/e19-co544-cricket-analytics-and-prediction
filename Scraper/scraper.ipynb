```python
import pandas as pd
import numpy as np

def combinedNames(people: pd.core.frame.DataFrame, names: pd.core.frame.DataFrame):
    #use identifier as index
	people.set_index('identifier', inplace=True)
	names.set_index('identifier', inplace=True)

	names_from_people_arr = people[['name', 'unique_name']].values
	names_from_names_arr = people.index.map(lambda x: names[names.index == x]['name'].values).values

	all_names : list[list[str]] = [None] * names_from_people_arr.shape[0]
	for i in range(names_from_people_arr.shape[0]):
		all_names[i] = np.unique(np.concatenate([names_from_people_arr[i], names_from_names_arr[i]]))

	people['all_names'] = all_names
	people.drop(columns=['name', 'unique_name'], inplace=True)
	people = people[['all_names', 'key_cricinfo']]
	return people

df = combinedNames(pd.read_csv("../Data/people.csv"), pd.read_csv("../Data/names.csv"))
```

# Web Scraper to Get Batsman/Bowler Type

```python
import requests
from bs4 import BeautifulSoup


class Player:
    def __init__(self, identifier, key_cricinfo, all_names):
        self.identifier = identifier
        self.key_cricinfo = key_cricinfo
        self.all_names = all_names
        self._urls = [
            f'https://www.espncricinfo.com/cricketers/{name.lower().replace(" ", "-")}-{key_cricinfo}'
            for name in all_names
        ]
        self.valid_url = None
        self.batting_style = None
        self.bowling_style = None
        self.soup = None  # Single attribute for parsed HTML

    def __str__(self):
        return f"PlayerURL(key_cricinfo={self.key_cricinfo}, all_names={self.all_names}, valid_url={self.valid_url})"

    def get_valid_url(self):
        """
        Fetches and stores the first valid URL from self.urls.

        Returns:
            str: The first valid URL from self.urls or None if none are valid.
        """
        if not self.valid_url:
            for url in self._urls:
                response = requests.get(url)
                if response.status_code == 200:
                    self.valid_url = url
                    self.soup = BeautifulSoup(response.content, 'html.parser')
                    break
        return self.valid_url

    def scrape_info(self):
        """
        Scrapes the valid URL (if available) to get player type information.

        Raises:
            ValueError: If no valid URL is found.

        Returns:
            tuple(str, str): A tuple containing batsman and bowler types (or None if not found).
        """

		# if no valid url, set player_type to None
        if not self.valid_url:
            print(f"No Valid URL Found for {self}")
            return 
	
	


        if not self.soup:
            # Fetch and parse HTML only if not already done
            response = requests.get(self.valid_url)
            response.raise_for_status()  # Raise exception for non-200 status codes
            self.soup = BeautifulSoup(response.content, 'html.parser')

        try:
            self.batting_style = self.soup.find(text="Batting Style").parent.parent.find_all("p")[1].text
            self.bowling_style = self.soup.find(text="Bowling Style").parent.parent.find_all("p")[1].text
        except (AttributeError, TypeError):
            self.batting_style = None
            self.bowling_style = None

        
```

```python
# For each player in df, create a PlayerURL object and fetch player type and add to df as new columns
df_with_types = df.iloc[15000:15037, ].copy()

# Using multithreading to speed up the process
from concurrent.futures import ThreadPoolExecutor

def get_player_info(index, row):
    player = Player(index, row['key_cricinfo'], row['all_names'])
    player.get_valid_url()
    player.scrape_info()
    return player

with ThreadPoolExecutor() as executor:
    players = list(executor.map(lambda x: get_player_info(*x), df_with_types.iterrows()))

for player in players:
	df_with_types.at[player.identifier, 'batting_style'] = player.batting_style
	df_with_types.at[player.identifier, 'bowling_style'] = player.bowling_style
```
